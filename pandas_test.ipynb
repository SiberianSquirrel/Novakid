```python
#parse json
with open('messages.json', encoding='utf-8') as json_file: 
    data = json.load(json_file) 
df_message=pd.json_normalize(data, 'conversation',[['participants']])
df_message.head()
```

```python
df_message=df_message[['sender', 'created_at', 'text','participants']]
df_message['created_at']=df_message['created_at']\
                                .apply(lambda x : datetime.strptime(x.split('+')[0].replace('T',' '),'%Y-%m-%d %H:%M:%S.%f'))
                                                            
df_message['date']=df_message['created_at'].apply(lambda x : x.date()) # –¥–∞—Ç–∞ —Å–æ–æ–±—â–µ–Ω–∏—è
df_message['sender_is_owner'] = np.where(df_message['sender']=='novakidschool', 'yes', 'no') # –º–µ—Ç–∫–∞ –≤—Ö–æ–¥—è—â–µ–≥–æ/–∏—Å—Ö–æ–¥—è—â–µ–≥–æ —Å–æ–æ–±—â–µ–Ω–∏—è
end_date=df_message['date'].max() # –º–µ—Ç–∫–∞ –æ—Ç –∫–æ—Ç–æ—Ä–æ–π —Å—á–∏—Ç–∞—Ç—å –∏–Ω—Ç–µ—Ä–≤–∞–ª
df_message['participants'] = df_message['participants'].apply(tuple) # —á—Ç–æ–± —Ä–∞–±–æ—Ç–∞–ª groupby
df_message.head()
```

```python
# 1) —Å–∫–æ–ª—å–∫–æ –±—ã–ª–æ –≤—Ö–æ–¥—è—â–∏—Ö —Å–æ–æ–±—â–µ–Ω–∏–π  –∑–∞ –ø–æ—Å–ª–µ–¥–Ω–∏–µ 3 –º–µ—Å.
df_message[(df_message['date'].between(end_date-pd.DateOffset(months=3),end_date ))
          & (df_message['sender_is_owner']=='no')].shape[0]
```

```python
#2) —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –ø–µ—Ä–≤–æ–≥–æ –æ—Ç–∫–ª–∏–∫–∞ (—Ç–æ –µ—Å—Ç—å –Ω–æ–≤–∞–∫–∏–¥ –æ—Ç–≤–µ—Ç–∏–ª –ø–µ—Ä–≤—ã–º) –ø–æ –¥–Ω—è–º - –ø–æ—Å–ª–µ–¥–Ω–∏–µ 30 —Å—É—Ç–æ–∫
df_message_last_thirty_days=df_message[(df_message['date'].between(end_date-pd.DateOffset(days=30),end_date ))] 

# –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è —Å–æ–æ–±—â–µ–Ω–∏—è –æ—Ç –Ω–æ–≤–∞–∫–∏–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é ,–≤ —Ä–∞–∑—Ä–µ–∑–µ –¥–∞—Ç
df_min_time_owner_message=df_message_last_thirty_days.loc[df_message_last_thirty_days['sender_is_owner']=='yes']\
                                                        .groupby(['participants','date'],as_index=False)['created_at']\
                                                        .min().rename(columns={'created_at':'min_owner_time'}) 

# –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è —Å–æ–æ–±—â–µ–Ω–∏—è –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –Ω–æ–≤–∞–∫–∏–¥—É  ,–≤ —Ä–∞–∑—Ä–µ–∑–µ –¥–∞—Ç
df_min_time_user_message=df_message_last_thirty_days.loc[df_message_last_thirty_days['sender_is_owner']=='no']\
                                                        .groupby(['participants','date'],as_index=False)['created_at']\
                                                        .min().rename(columns={'created_at':'min_user_time'})

# —Ä–µ–∑—É–ª—å—Ç–∏—Ä—É—é—â–∏–π –¥—Ñ –ø–æ –¥–∞—Ç–µ –∏ —É—á–∞—Å—Ç–Ω–∏–∫–∞–º –¥–∏–∞–ª–æ–≥–∞
df_min_time=pd.merge(df_min_time_owner_message,df_min_time_user_message,how='inner',on=['participants','date'])

# –±–µ—Ä–µ–º —Å–ª—É—á–∞–∏, –∫–æ–≥–¥–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –Ω–∞–ø–∏—Å–∞–ª –ø–µ—Ä–≤—ã–º –∑–∞ –¥–µ–Ω—å
df_min_time=df_min_time.loc[df_min_time['min_owner_time']>df_min_time['min_user_time']] 

# –≤—Ä–µ–º—è –æ—Ç–∫–ª–∏–∫–∞ –≤ –º–∏–Ω—É—Ç–∞—Ö
df_min_time['difference_per_minute']=(df_min_time['min_owner_time']-df_min_time['min_user_time']).apply(lambda x: x.total_seconds()/60) 

# —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –æ—Ç–∫–ª–∏–∫–∞ –ø–æ –¥–Ω—è–º 
df_min_time.resample('d', on='min_user_time')['difference_per_minute'].mean().to_frame().T

#df_message_last_thirty_days.loc[df_message_last_thirty_days['participants']==('balzh_5', 'novakidschool')]
#df_message_last_thirty_days['next_datetime']=df_message_last_thirty_days.groupby(['participants','date'])['created_at'].shift(1)
```

```python
#3) –∫–∞–∫–æ–π –ø—Ä–æ—Ü–µ–Ω—Ç —à–∞–±–ª–æ–Ω–Ω—ã—Ö —Ñ—Ä–∞–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç–µ–ª–∏ –Ω–æ–≤–∞–∫–∏–¥ –∑–∞ –ø–æ—Å–ª–µ–¥–Ω–∏–π –ø–æ–ª–Ω—ã–π –º–µ—Å—è—Ü?. —Å–ø–∏—Å–æ–∫ —Ñ—Ä–∞–∑ —Å–º –Ω–∏–∂–µ.
list_found_patterns=df_message[(df_message['date'].between(end_date-pd.DateOffset(months=1),end_date ))
           & (df_message['sender_is_owner']=='yes')]\
            ['text'].apply(lambda x : [pattern  for pattern in  phrases_list if pattern in str(x)]).to_list() # –ø—Ä–æ–≤–µ—Ä–∫–∞ –≤—Ö–æ–∂–¥–µ–Ω–∏–π —Ñ—Ä–∞–∑ –≤ —Å–æ–æ–±—â–µ–Ω–∏–µ

list_found_unique_patterns=list(set([ i for j in list_found_patterns for i in j])) # —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ø–∏—Å–æ–∫ –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö —Ñ—Ä–∞–∑ (–∏–∑–±–∞–≤–ª–µ–Ω–∏–µ –æ—Ç –≤–ª–æ–∂–µ–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞)
print(f'–ü—Ä–æ—Ü–µ–Ω—Ç–æ–≤ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–Ω—ã—Ö —à–∞–±–ª–æ–Ω–Ω—ã—Ö —Ñ—Ä–∞–∑ {round(len(list_found_unique_patterns)/len(phrases_list)*100,2)} %')
#['–í—Ä–µ–º—è –∏ –¥–∞—Ç—É –ø—Ä–æ–±–Ω–æ–≥–æ —É—Ä–æ–∫–∞',
# '–í—Ä–µ–º—è –∏ –¥–∞—Ç—É –ø—Ä–æ–±–Ω–æ–≥–æ —É—Ä–æ–∫–∞ –º–æ–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å –ø–æ —Å—Å—ã–ª–∫–µ']
```

```python
#4) –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö –æ—Ç novakidschool / –∏ –Ω–æ–≤–∞–∫–∏–¥—É —Å–æ–æ–±—â–µ–Ω–∏–π –∑–∞ –Ω–µ–¥–µ–ª—é - –ø–æ—Å–ª–µ–¥–Ω–∏–µ 10 –Ω–µ–¥–µ–ª—å.
df_message[(df_message['date'].between(end_date-pd.DateOffset(weeks=10),end_date ))]\
              .groupby('sender_is_owner').resample('w', on='created_at').size()
              
```

```python
#5) –¥–∏–∞–ª–æ–≥–æ–≤ –∑–∞ –ø–æ—Å–ª–µ–¥–Ω–∏–µ 3 –º–µ—Å—è—Ü–∞ –ø–æ–º–µ—Å—è—á–Ω–æ –Ω–∞ –∫–∞–∫–∏–µ –ù–æ–≤–∞–∫–∏–¥ –Ω–µ –ø–æ–ª—É—á–∏–ª –æ—Ç–≤–µ—Ç–∞
df_message_last_three_months=df_message[(df_message['date'].between(end_date-pd.DateOffset(months=3),end_date ))]

# –ø–æ–¥—Å—á–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –æ—Ç–ø—Ä–∞–≤–∏—Ç–µ–ª–µ–π –ø–æ —É—á–∞—Å—Ç–Ω–∏–∫–∞–º –¥–∏–∞–ª–æ–≥–æ–≤
df_message_last_three_months['count_sender']=df_message_last_three_months.groupby(['participants'])['sender'].transform('nunique')
#df_message_last_three_months.loc[df_message_last_three_months['participants']==('novakidschool', 'soulinpeony')]

# –±–µ—Ä–µ–º –¥–∏–∞–ª–æ–≥–∏ –≥–¥–µ –æ—Ç–ø—Ä–∞–≤–∏—Ç–µ–ª—å –±—ã–ª —Ç–æ–ª—å–∫–æ –Ω–æ–≤–∞–∫–∏–¥
df_message_last_three_months[(df_message_last_three_months['count_sender']==1)&
                            (df_message_last_three_months['sender_is_owner']=='yes')]\
                            .resample('m', on='created_at').nunique()['participants'].to_frame()
                             # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –¥–∏–∞–ª–æ–≥–æ–≤ –≤ —Ä–∞–∑—Ä–µ–∑–µ –º–µ—Å—è—Ü–µ–≤
```
